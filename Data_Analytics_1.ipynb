# Data Mining vs Knowledge Discovery from Data(KDD)
## Data Mining and KDD are often used interchangeably. But it can also be understood by some slight difference.
### <i>Note 1 :-</i> Data Mining is the use of algorithms to extract the information and patterns designed by the KDD process.
### <i>Note 2 :-</i> KDD is the process of finding the useful information and patterns in data.
### All in all Data Mining is a intermediate step in the KDD process.
